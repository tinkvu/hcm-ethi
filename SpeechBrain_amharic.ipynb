````python
import torch
from speechbrain.inference.ASR import EncoderASR

# Load the model
asr_model = EncoderASR.from_hparams(source="speechbrain/asr-wav2vec2-dvoice-amharic", savedir="pretrained_models/asr-wav2vec2-dvoice-amharic")

# Perform dynamic quantization
asr_model.eval()  # Make sure the model is in evaluation mode
quantized_model = torch.quantization.quantize_dynamic(asr_model, dtype=torch.qint8)

# Use quantized model for transcription
result = quantized_model.transcribe_file('speechbrain/asr-wav2vec2-dvoice-amharic/example_amharic.wav')
print(result)
````

````python
import time
start=time.time()
transcription=quantized_model.transcribe_file('/content/Interview 002.mp4')
end=time.time()
print("Time taken: ", end-start)
transcription
````

````python
from torch.nn.utils import prune
import torch
from speechbrain.inference.ASR import EncoderASR

# Load the model
asr_model = EncoderASR.from_hparams(source="speechbrain/asr-wav2vec2-dvoice-amharic", savedir="pretrained_models/asr-wav2vec2-dvoice-amharic")

# Perform pruning (remove 20% of the model's weights)
asr_model.eval()  # Make sure the model is in evaluation mode
for name, module in asr_model.named_modules():
    if isinstance(module, torch.nn.Linear):
        prune.random_unstructured(module, name="weight", amount=0.2)

# Use pruned model for transcription
result = asr_model.transcribe_file('speechbrain/asr-wav2vec2-dvoice-amharic/example_amharic.wav')
print(result)
````
